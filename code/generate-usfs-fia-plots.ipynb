```python
import matplotlib.pyplot as plt
```

```python
# USFS-FIA coconut palm population estimates for Guam
x = [2002, 2013]
y = [1162494, 1464e3]
se = [None, 528e3]
```

```python
# Existing scatter plot data
x_scatter = x
y_scatter = y

# Create scatter plot
plt.plot(x_scatter, y_scatter, 'o:', color='blue', markersize=10)

# Draw a line to indicate sampling error for the 2013 data
x_line = [x[1], x[1]]
y_line = [y[1] - se[1], y[1] + se[1]]
plt.plot(x_line, y_line, color='blue', linewidth=3)

# Show the plot
plt.xlabel('survey date')
plt.ylabel('number of coconut palms (millions)')
plt.title('Population estimates of coconut palms on Guam by USFS-FIA')
plt.grid(False)
plt.ylim(0, 2.1e6)
plt.annotate(
    text='Oryctes rhinoceros\nfirst detected',
    xy=(2007.75, 0), 
    xytext=(2007.75, 0.3e6),
    arrowprops=dict(arrowstyle='->', 
    color='black'),
    fontsize=12,
    ha='center')
plt.show()
# plt.savefig('../images/coconut_palm_population_guam.png', dpi=300, bbox_inches='tight')
```

Island-wide population estimates for Guam estimated by the US Forest Service from Forest Inventory Assessment surveys performed in 2002 and 2013. 

Note that different metrics were used in the two surveys. 

The 2002 estimate comes from [Donnegon et al. 2004](http://www.fs.fed.us/pnw/pubs/pnw_rb243.pdf), page 16, Table 5: *Estimated number of live trees on forest land by species and diameter class*. They reported a total of 1,162,494 live coconut palms growing on forest land on Guam during 2002.

The 2013 estimate comes from [Lazaro et al. 2020](http://www.fs.usda.gov/treesearch/pubs/59433), page 7, Tabel 2: *Estimated number of trees on Guam, by species*. They simply reported their estimate of the total number of coconut palms on Guam during 2013 as 1,464,000 with a sampling error of 528,000.

They reported a total of 1,571,556 live *Cycas micronesica* plamts growing on forest land on Guam during 2002.
